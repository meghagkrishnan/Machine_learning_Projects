### Used transfer learning to predict Tomato Leaf Disease 
### Used pre-trained model from google's Tensorflow Hub ([TF2] Imagenet (ILSVRC-2012-CLS) classification with MobileNet V2) and re-train that on Tomato leaf dataset from Kaggle.
### Dataset Link: https://www.kaggle.com/datasets/charuchaudhry/plantvillage-tomato-leaf-dataset


```python
import tensorflow as tf
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
import PIL.Image as Image
import numpy as np
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "Tomato",
    image_size = (256,256),
    batch_size = 32
    )
```

```python
def data_partition(ds, train_size,test_size,shuffle=True, shuffle_size=1000):
    
    ds=dataset
    
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12) #if shuffle option is given in function. seed can be any number to make a constant shuffle
    
    train_data = dataset.take(int(len(dataset)*train_size))
    remain_data = dataset.skip(len(train_data))
    test_data = dataset.take(int((len(dataset)*test_size)/2))
    val_data = remain_data.skip(len(test_data))
    
    return train_data, test_data, val_data
```

```python
train_data, test_data, val_data =data_partition(dataset, 0.8, 0.2,shuffle=True, shuffle_size=1000)
```

```python
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(224,224), #If any input image is of not this size
    layers.experimental.preprocessing.Rescaling(1.0/255)
])
```

```python
data_augmentation = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"), #If any input image is of not this size
    layers.experimental.preprocessing.RandomRotation(0.2)
    
])
```

```python
feature_extractor_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

pretrained_model_without_top_layer = hub.KerasLayer(
    feature_extractor_model, input_shape=(224, 224, 3), trainable=False) # trainable=False to freeze the hidden layers
```

```python
n_classes = 10

model = tf.keras.Sequential([
    resize_and_rescale,
    data_augmentation,
    pretrained_model_without_top_layer,
    tf.keras.layers.Dense(n_classes)
])

#model.summary()
```

```python
model.compile(
  optimizer="adam",
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['acc'])
```

```python
history = model.fit(
    train_data,
    batch_size=32,
    validation_data=val_data,
    verbose=1,
    epochs=3,
)
```

#### Within 3 epochs, 87% accuracy is obtained

```python
model.evaluate(train_data)
```
