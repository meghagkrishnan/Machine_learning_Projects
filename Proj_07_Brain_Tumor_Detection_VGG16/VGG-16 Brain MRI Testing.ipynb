```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "BrainMRI",
    image_size = (256,256),
    batch_size = 5
    )
```

```python
class_names = dataset.class_names
class_names
```

```python
imgs=[]
labls=[]

for image_batch, label_batch in dataset:
    for images in image_batch:
        imgs.append(images)
    for labels in label_batch:
        labls.append(labels)
    
```

```python
from keras.models import load_model
saved_model = load_model("model.h5")
```

```python
n=np.random.randint(0, 50)
img = imgs[n].numpy().astype("uint8")
plt.imshow(img)

img = np.expand_dims(img, axis=0)
prediction = np.argmax(saved_model.predict(img))
    
print("The prediction for this image is: ", class_names[prediction])
print("The actual label for this image is: ", class_names[labls[n]])
```
